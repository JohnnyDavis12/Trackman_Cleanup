```python
#Importing the libraries
import pandas as pd
import pickle
from pandas import DataFrame, Series
import numpy as np
```

```python
# Read in the data / TrackMan Data
df = pd.read_csv("/Users/johndavis/Desktop/August_1.csv",low_memory=False)
df.head(25)
```

```python
#Create a new dataframe with only the columns we want for Analysis
dft = df[['Pitcher','RelSpeed','SpinRate','RelHeight','RelSide','Extension','InducedVertBreak', 'HorzBreak']]
#Rename the columns to be more descriptive
dft.rename(columns={'RelSpeed':'Velocity', 'SpinRate':'Spin Rate', 'RelHeight':'Release Height', 'RelSide':'Release Side', 'Extension':'Extension', 'InducedVertBreak':'Induced Vertical Break', 'HorzBreak':'Horizontal Break'}, inplace=True)
```

```python
#Add the Pitch Type and Pitch Result to the new dataframe
dft.insert(1, 'Pitch Type', df['AutoPitchType'])
dft.insert(2, 'Pitch Result', df['PitchCall'])
#Create a new column for the difference between the Induced Vertical Break and the Horizontal Break (Absolute Value)
dft.assign(Differential_Break = dft['Induced Vertical Break'] - dft['Horizontal Break'].abs())
#Add two new columns for Pitch Location
dft.insert(10, 'Pitch Height', df['PlateLocHeight'])
dft.insert(11, 'Pitch Side', df['PlateLocSide'])
```

```python
#Delete all rows with NaN values
dft.dropna(inplace=True)
```

```python
#Output the new dataframe to a csv file
dft.to_csv('August_1_Updated.csv')
```
